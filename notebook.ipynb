```python
# Re-run this cell
# Preloaded data for validating email domain.
top_level_domains = [
    ".org",
    ".net",
    ".edu",
    ".ac",
    ".gov",
    ".com",
    ".io"
]
```

```python
def validate_name(name: str):     
    if isinstance(name, str):        
        if len(name) > 2:             
            return True         
        else:             
            return False     
    else:         
        return False  
def validate_email(email: str):
    if '@' in email:
        for tld in top_level_domains:
            if email.endswith(tld):
                return True
        return False
    else:   
        return False
        
       

        
```
